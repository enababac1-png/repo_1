```python
def ph_evaluator(c=0.001):
  '''Ph decision maker'''
  ph_value = ph(c)

  if ph_value < 7:
    decision = 'kiselo'
  else:
    decision = 'lunato'

  return decision


ph_evaluator()
```

```python
def ph_evaluator(c=0.001):
  if ph(c) < 7:
    return 'kiselo'
  else:
    return 'lunato'


conc_list = [0.001, 0.01, 0.1, 1, 10]
for solution in conc_list:
    print(ph_evaluator(solution))
```
