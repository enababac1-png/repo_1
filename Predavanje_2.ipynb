```python
#brišemo sve retke i stupce koji imaju vrijednosti NaN, ostale zamjenimo sa 0.001

df[b_df].dropna(how='all', axis=0).dropna(how='all', axis=1).dropna(how='any', axis=0).dropna(how='any', axis=1)
```

```python
#popuniti stupac Z s nulama
#na cijeli df se primjeni if else funkcija
#ako je vrijednost veća od 0.01, ispiši tu vrijednost, ako nije onda 0.001

df['Z'] = 0
df = df.applymap(lambda x: x if x > 0.01 else 0.001)
df
```

```python
#ukloni stupac Z iz df
#zamjeni 0.001 s NaN
#ukloni stupce koji imaju vrijednosti NaN, a preostale ponovo zamjeni sa 0.001

df.drop('Z', axis=1)
df.replace(0.001, np.nan).dropna(how='all', axis=1).replace(np.nan, 0.001)
```
